```python
import numpy as np
import matplotlib.pyplot as plt
```

# オイラー法（一変数）
$$
x_{n+1} = x_n + \varDelta t f(x,t)
$$

# 例：ロジスティック方程式

$$
\frac{dx}{dt} = r x (1- \frac{x}{K})
$$

```python
# パラメータ
r = 0.5 # 増殖率
K = 200 # 環境収容力

# ロジスティック方程式（の右辺）
def f(x):
    return r*x*(1- x/K)

# 時間や変数の配列の設定
dt = 0.01 # 刻み幅
t0 = 0.0  # 初期時刻
tf = 20.0 # 終了時刻
t = np.arange(t0, tf, dt) # 時刻の配列
x = np.zeros(len(t)) # 個体数の配列

# 初期値
x[0] = 2

# オイラー法
for i in range(len(t)-1):
    x[i+1] = x[i] + dt*f(x[i])

# グラフの描画
plt.plot(t, x)
plt.xlabel("Time")
plt.ylabel("Population")
plt.show()
```
