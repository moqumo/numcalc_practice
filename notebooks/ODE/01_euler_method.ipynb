# オイラー法（一変数）
常微分方程式方程式
$$
\frac{dx}{dt} = f(x,t)
$$
の数値計算を考える．
### 結論 
$x_n = x\big(n\varDelta t\big)~(n=1,2,3..., \varDelta t<<1)$とすると，
$$
x_{n+1} = x_n + \varDelta t f(x,t)
$$
と逐次的に解を求められる．
### 図解
![](images/euler.png)

### 簡単な導出
微分の定義から
$$
\frac{dx}{dt} = \lim_{\varDelta t \to 0}\frac{x(t + \varDelta t)- x(t)}{\varDelta t}
$$
より，$\varDelta t<<1$とすると
$$
\frac{dx}{dt} \simeq \frac{x(t + \varDelta t)- x(t)}{\varDelta t}
$$
となる．ここで$\frac{dx}{dt} = f(x,t)$であるから，
$$
\frac{x(t + \varDelta t)- x(t)}{\varDelta t} \simeq f(x,t)
$$
したがって，$x(t + \varDelta t)$を左辺に持ってくるように変形すると，
$$
x(t + \varDelta t) \simeq x(t) + \varDelta t f(x,t)
$$

# 例：ロジスティック方程式

個体数$x=x(t)$
$$
\frac{dx}{dt} = r x (1- \frac{x}{K})
$$
$r$:増殖率   
$K$:環境収容力   

```python
# パラメータ
r = 0.5 # 増殖率
K = 200 # 環境収容力

# ロジスティック方程式（の右辺）
def f(x):
    return r*x*(1- x/K)

# 時間や変数の配列の設定
dt = 0.01 # 刻み幅
t0 = 0.0  # 初期時刻
tf = 20.0 # 終了時刻
t = np.arange(t0, tf, dt) # 時刻の配列
x = np.zeros(len(t)) # 個体数の配列

# 初期値
x[0] = 2

# オイラー法
for i in range(len(t)-1):
    x[i+1] = x[i] + dt*f(x[i])

# グラフの描画
plt.plot(t, x)
plt.xlabel("Time")
plt.ylabel("Population")
plt.grid()
plt.show()
```
